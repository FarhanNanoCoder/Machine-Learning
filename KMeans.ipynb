```python
def ecl (p1,p2):
    return np.sqrt(np.sum((p1-p2)**2))
```

```python

def print_graph(datapoint,label,cen,k):
    for i in range(k):
        points=datapoint[label==i]
        plt.scatter(points[:,0],points[:,1],label="Cluster: "+str(i+1))
        plt.scatter(cen[0][i],cen[i][1],label="Centroid: "+str(i+1))
    plt.legend()
    plt.show()
```

```python
def k_means(datapoint,k,epoch=10):
    cen=np.array([(np.random.randint(0,20),np.random.randint(0,20)) for i in range (k)])
    print('cen',cen)
    for e in range (epoch):
        label=[]  #holds the closest centroid index for each datapoint
        for d in datapoint:
            min_idx = np.argmin([ecl(d,cen[c]) for c in range (k)])
            label.append(min_idx)
        label = np.array(label)

        print_graph(datapoint,label,cen,k)

        for c in range (k):
            cen[c] = np.mean(datapoint[label==c],axis=0)   
    
    return cen
```
