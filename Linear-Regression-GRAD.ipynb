```python
import numpy as np
import matplotlib.pyplot as plt
```

## Data Handlation & Visuaization

```python
m = 1.2
c = 2
X = np.array([i for i in range(-1, 50)])
Y_OLD = m*X + c
Y = Y_OLD + np.random.normal(0, 3, X.shape[0])

plt.plot(X,Y_OLD,label='Actual', color='blue')
plt.scatter(X,Y, color='red',label='Predicted')
plt.legend()
plt.show()
```

## Model, Loss Function & Optimizer

```python
def model (X, w, b):
    return w*X + b
```

```python
def loss(Y,Y_):
    return np.sum((Y-Y_)**2)/Y.shape[0]
```

```python
def grad_desc(X,Y,w=0.6,b=5,epoch=10,lr=0.01):
    losses=[]
    ws=[]

    for e in range(epoch):
        Y_ = model(X,w,b)
        l = loss(Y,Y_)
        losses.append(l)
        ws.append(w)
        dw = np.sum(2*(Y_-Y)*X)/Y.shape[0]
        db = np.sum(2*(Y_-Y))/Y.shape[0]

        print('Epoch: ',e+1,' Loss: ',l,' w: ',w,' b: ',b)
        plt.plot(X,Y_,label="Epoch: "+str(e+1))
        w = w - lr*dw
        b = b - lr*db

    plt.scatter(X,Y,label='Actual')
    plt.legend()
    plt.show()

    return ws,losses
```

## Test

```python
grad_desc(X,Y,epoch=10,lr=0.001)
```
